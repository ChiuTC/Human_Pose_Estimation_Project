### 定義testing data的路徑, 定義main_test函式，用來評估testing data

```python
# Level1
# Hand waving
video_hand = np.array(['testing/1/S001C001P006R002A023_rgb.avi',
                       'testing/1/S001C001P007R001A023_rgb.avi'])
# Kicking something
video_kick = np.array(['testing/1/S001C001P006R002A024_rgb.avi',
                       'testing/1/S001C001P007R001A024_rgb.avi'])
# Level2
# Sit down
video_sit_down = np.array(['testing/2/S001C001P006R002A008_rgb.avi',
                           'testing/2/S001C001P007R001A008_rgb.avi'])
                    
# Stand up
video_stand_up = np.array(['testing/2/S001C001P006R002A009_rgb.avi',
                           'testing/2/S001C001P007R001A009_rgb.avi'])

# Level3
# Reading 
video_reading  = np.array(['testing/3/S001C001P006R002A011_rgb.avi',
                           'testing/3/S001C001P007R001A011_rgb.avi'])
# Writing
video_writing  = np.array(['testing/3/S001C001P006R002A012_rgb.avi',
                           'testing/3/S001C001P007R001A012_rgb.avi'])
# Play with phone
video_play = np.array(['testing/3/S001C001P006R002A029_rgb.avi',
                       'testing/3/S001C001P007R001A029_rgb.avi'])
cal_time = np.array([])

def main_test():
    global cal_time
    cal_time = np.array([])
    video_test = np.array([])
    # Level1
    start = time.time()
    for addr in video_hand:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
            
    for addr in video_kick:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
    end = time.time()  
    cal_time = np.append(cal_time,(end - start))
    # Level2      
    start = time.time()
    for addr in video_sit_down:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
            
    for addr in video_stand_up:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
    end = time.time()
    cal_time = np.append(cal_time,(end - start))
    # Level3
    start = time.time()
    for addr in video_reading :
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
            
    for addr in video_writing:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
            
    for addr in video_play:
        cap = cv2.VideoCapture(addr) 
        is_level3 = myFunctions.is_leve1Orlevel2(cap)
        if(is_level3):
            video_test = np.append(video_test, addr) 
#     print(video_test)
    myFunctions.detectLevelThree(video_test)
    end = time.time()
    cal_time = np.append(cal_time,(end - start))
```
